```python
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from scipy import ndimage
import cv2
import random
```

```python
PATH = 'chest_xray/'
```

```python
# check class balance for training set
train_path = f'{PATH}train/'
normal_list = os.listdir(f'{train_path}NORMAL')
sick_list = os.listdir(f'{train_path}PNEUMONIA')

plt.figure(figsize=(5, 3))
plt.bar(0, len(normal_list), tick_label='Normal')
plt.bar(1, len(sick_list), tick_label='Pneumonia')
plt.title('Class Balance in Training Set')
plt.ylabel('# of Pics')
plt.xlabel('Patient Type')
plt.show()
```

```python
# check class balance for testing set
test_path = f'{PATH}test/'
normal_list = os.listdir(f'{test_path}NORMAL')
sick_list = os.listdir(f'{test_path}PNEUMONIA')

plt.figure(figsize=(5, 3))
plt.bar(0, len(normal_list), tick_label='Normal')
plt.bar(1, len(sick_list), tick_label='Pneumonia')
plt.title('Class Balance in Testing Set')
plt.ylabel('# of Pics')
plt.xlabel('Patient Type')
plt.show()
```

```python
# Opening a random image from each category in training set

# From Normal category
normal_path = f'{train_path}NORMAL/'
im_list = os.listdir(normal_path)
im_path = f'{normal_path}{random.choice(im_list)}'
image = cv2.imread(im_path)

plt.imshow(image)
print(image.shape)
```

```python
# From Pneumonia category
sick_path = f'{train_path}PNEUMONIA/'
im_list2 = os.listdir(sick_path)
im_path = f'{sick_path}{random.choice(im_list2)}'
image = cv2.imread(im_path)

plt.imshow(image)
print(image.shape)
```

```python
# Inspecting the dimension of training set images

widths = []
heights = []
for images in im_list:
    if images == '.DS_Store':
        continue
    im_path = f'{normal_path}{images}'
    image = Image.open(im_path)
    size = image.size
    widths.append(size[0])
    heights.append(size[1])

for images in im_list2:
    if images == '.DS_Store':
        continue
    im_path = f'{sick_path}{images}'
    image = Image.open(im_path)
    size = image.size
    widths.append(size[0])
    heights.append(size[1])

plt.hist(widths)
plt.title("Widths of X-Rays")
plt.ylabel('# of Pics')
plt.xlabel('Image Width')
plt.show()

plt.hist(heights)
plt.title("Heights of X-Rays")
plt.ylabel('# of Pics')
plt.xlabel('Image Height')
plt.show()
```
